## Amoeba population

There is a one amoeba in a pond. After every minute the amoeba may die, stay the same, split into two or split into three with equal probability. All its offspring, if it has any, will behave the same (and independent of other amoebas).

#### What is the probability the amoeba population will die out?

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
num_ponds = 10000   # Total number of ponds
num_steps = 10    # Total number of minutes
population_cap = 100000  # Maximum population per pond

# Each pond starts with one amoeba, stored as a 64-bit integer.
populations = np.ones(num_ponds, dtype=np.int64)

# For tracking the number of extinct ponds over time
extinct_over_time = []

# Run the simulation for each minute
for step in range(num_steps):
    new_populations = np.zeros(num_ponds, dtype=np.int64)
    for i in range(num_ponds):
        # Current number of amoebae in the pond
        N = populations[i]
        if N == 0:
            # If the pond is already extinct, keep it that way.
            new_populations[i] = 0
        else:
            # For each of the N amoebae, simulate the reproduction outcome:
            # Outcome mapping: 0 (die), 1 (stay the same), 2 (split into two), 3 (split into three)
            outcomes = np.random.multinomial(N, [0.25, 0.25, 0.25, 0.25])
            new_population = outcomes[1] + 2 * outcomes[2] + 3 * outcomes[3]
            # Cap the population at population_cap
            new_populations[i] = min(new_population, population_cap)
    
    populations = new_populations
    extinct_over_time.append(np.sum(populations == 0))
    
    # Optional: Print progress every 100 steps
    if (step + 1) % 100 == 0:
        print(f"Step {step+1}/{num_steps}: {np.sum(populations == 0)} ponds extinct.")

print("\nSimulation complete.")
print("Final populations for the first 10 ponds:", populations[:10])
print("Total number of extinct ponds:", np.sum(populations == 0))

# Plot histogram of the pond populations.
plt.figure(figsize=(10, 6))
plt.hist(populations, bins=50, edgecolor='black')
plt.xlabel('Population size')
plt.ylabel('Number of ponds')
plt.title('Histogram of Pond Populations after 1000 Minutes')
plt.show()

# Plot the number of extinct ponds over time.
plt.figure(figsize=(10, 6))
plt.plot(extinct_over_time)
plt.xlabel('Time step (minute)')
plt.ylabel('Number of extinct ponds')
plt.title('Extinct Ponds Over Time')
plt.show()
```
